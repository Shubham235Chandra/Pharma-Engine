```python
data['medicine_id'] = range(1, len(data) + 1)
data['manufacturer_id'] = data['product_manufactured'].astype('category').cat.codes + 1
data.head()
```

```python
# Medicine Table
medicine_table = data[['medicine_id', 'product_name', 'sub_category', 'salt_composition', 'medicine_desc']].copy()
```

```python
# Manufacturer Table
manufacturer_table = data[['manufacturer_id', 'product_manufactured']].drop_duplicates().copy()
```

```python
# Price Table
price_table = data[['medicine_id', 'product_price']].copy()
price_table['price_id'] = range(1, len(price_table) + 1)
```

```python
# Side Effects Table
side_effects_table = data[['medicine_id', 'side_effects']].copy()
side_effects_table['side_effect_id'] = range(1, len(side_effects_table) + 1)
```

```python
# Drug Interactions Table
# The drug_interactions column contains JSON-like strings. We need to normalize this into a separate DataFrame.
interaction_list = []
for idx, row in data.iterrows():
    try:
        interactions = json.loads(row['drug_interactions'].replace("'", '"'))
        interactions['medicine_id'] = row['medicine_id']
        interaction_list.append(interactions)
    except Exception as e:
        print(f"Error in row {idx}: {e}")
```

```python
interaction_list[0:3]
```

```python
drug_interactions_table = pd.json_normalize(interaction_list, 'drug', ['medicine_id'])
drug_interactions_table.columns = ['drug_interactions', 'medicine_id']
drug_interactions_table['inte/raction_id'] = range(1, len(drug_interactions_table) + 1)
```

```python
# Medicine_Manufacturer Association Table
medicine_manufacturer_table = data[['medicine_id', 'manufacturer_id']].copy()
medicine_manufacturer_table['association_id'] = range(1, len(medicine_manufacturer_table) + 1)
```

```python
import sqlite3
import pandas as pd

# Create a new SQLite database or connect to an existing one
db_file = 'medicine_database.db'
conn = sqlite3.connect(db_file)

# Define the dictionaries
table_dict = {
    'Medicine Table': medicine_table,
    'Manufacturer Table': manufacturer_table,
    'Price Table': price_table,
    'Side Effects Table': side_effects_table,
    'Drug Interactions Table': drug_interactions_table,
    'Medicine_Manufacturer Table': medicine_manufacturer_table,
}

# Iterate through the dictionary and create tables in the database
for table_name, df in table_dict.items():
    df.to_sql(table_name, conn, if_exists='replace', index=False)

# Commit changes and close the database connection
conn.commit()
conn.close()
```
